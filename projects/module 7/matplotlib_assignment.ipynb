# Matplotlib Assignment

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Sample Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, width=0.5, color='green')
plt.title('Student Marks')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.xticks(rotation=45)
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = (0.1, 0, 0, 0)  # only "explode" the 1st slice (North America) as it has the highest revenue
plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode, startangle=140)
plt.title('Revenue Distribution by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
np.random.seed(0)  # For reproducibility
data = np.random.randint(1, 101, size=1000)
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
